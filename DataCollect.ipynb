```python
import json


filename = "buzzfeed-v02.txt"
dict1 = {}
dict2 = {}
fields =['text','label']
with open(filename) as tt:
    file = tt.read()
    idx = 0
    for instance in file.split('\n'):
        dict1[fields[0]] = instance.split('\t',6)[3]
        dict1[fields[1]] = 0
        emu = 'train-'+str(idx)
        dict2[emu] = dict(dict1)
        idx = idx + 1

out_file = open("buzzfeed.json", "w") 
json.dump(dict2, out_file) 
out_file.close() 
```

```python
import json
def get_data_from_file(filepath,ty):
    with open(filepath) as jsonfile:
        data = json.load(jsonfile)
    sentences = []
    labels = []
    size = len(data)
    for i in range(size):
        sentences.append(data[ty+'-'+str(i)]['text'])
        if'label' in data[ty+'-'+str(i)].keys():
            labels.append(data[ty+'-'+str(i)]['label'])
    return sentences,labels 


text,label = get_data_from_file("buzzfeed.json","train")
```
